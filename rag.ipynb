```python
def text_formatting(text:str) -> str:
    cleaned_text=text.replace("\n", " ")
    return cleaned_text

def open_and_read_pdf(pdf_path: str) -> list[dict]:
    doc = fitz.open(pdf_path)
    page_and_text=[]
    for page_number,page in tqdm(enumerate(doc)):
        text = page.get_text()
        text = text_formatting(text)
        page_and_text.append({"page_number": page_number, "page_char_count": len(text), "word_count":len(text.split()), "sentence_count": len(text.split(".")),"page_token":len(text)/4, "text": text})
    return page_and_text

page_and_text = open_and_read_pdf("hso.pdf")
page_and_text[3]
```

## Chunking our sentence together

```python
import pandas as pd
from tqdm import tqdm

# --- Helper function to group sentences into chunks ---
def create_chunks(sentences, chunk_size=3, overlap=1):
    """
    Groups a list of sentences into overlapping chunks of text.
    Example:
        sentences = ["s1", "s2", "s3", "s4", "s5"]
        chunk_size=3, overlap=1 -> ["s1 s2 s3", "s3 s4 s5"]
    """
    if overlap >= chunk_size:
        raise ValueError("Overlap size must be smaller than chunk size.")
        
    chunks = []
    step = chunk_size - overlap
    for i in range(0, len(sentences), step):
        chunk_sentences = sentences[i:i + chunk_size]
        if not chunk_sentences:
            continue
        chunks.append(" ".join(chunk_sentences))
    return chunks

# --- Add chunks back into your page_and_text list of dicts ---
for item in tqdm(page_and_text):
    sentences = item.get("sentences", [])
    item["chunks"] = create_chunks(sentences, chunk_size=3, overlap=1)


#df.chunks[0]
df.head(10)
```

## Merge chunk based on simillarity

```python
final_sentences=[]
for item in tqdm(page_and_text):
    for sentence in item["sentences"]:
        final_sentences.append(sentence)
print(len(final_sentences))
print (final_sentences[0:5])
```

```python
from sentence_transformers import SentenceTransformer
from tqdm import tqdm
import numpy as np
import uuid

# --- 1. Load the model ONCE ---
model = SentenceTransformer("BAAI/bge-m3", device="cpu")  # change to "cuda" if you have GPU

# --- 2. Compute embeddings for all texts ---
embeddings = []
for item in tqdm(final_sentences):  # 'text' should be a list of strings
    emb = model.encode(item, convert_to_numpy=True)
    embeddings.append( emb)


```

## Cluster the chunk symmentially

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# suppose you already have embeddings as a numpy array (num_sentences, 1024)|


final_chunks = []
curr_chunk = final_sentences[0]  # start with the first sentence

# iterate through embeddings starting from the second one
for i in range(1, len(embeddings)):
    if embeddings[i].shape != (1024,):
        print(f"Warning: Embedding at index {i} has shape {embeddings[i].shape}, expected (1024,)")
    else:
        # cosine similarity between consecutive sentences
        sim = cosine_similarity([embeddings[i-1]], [embeddings[i]])[0][0]
        
        if sim > 0.36:  # threshold for high similarity
            curr_chunk = " ".join([curr_chunk, final_sentences[i]])  # append to current chunk
        else:
            final_chunks.append(curr_chunk)  # save the completed chunk
            curr_chunk = final_sentences[i]  # start a new chunk

# append last chunk
final_chunks.append(curr_chunk)


```

```python
print("Number of final chunks:", len(final_chunks))
print("Sample chunks:", final_chunks[1])
#print("length of sample chunk:", len(final_chunks[0]))
```

## make database to store embedding part

```python
# embedd all final chunks
chunks_embeddings=[]
for chunk in final_chunks:
    emb = model.encode(chunk, convert_to_numpy=True)
    chunks_embeddings.append( emb)
```

```python
chunks_embeddings = np.array(chunks_embeddings)
print(chunks_embeddings.shape)   # (num_chunks, embedding_dim)  # number of dimensions of chunks_embeddings
print(len(final_chunks))              # length of final_chunks list
print(type(chunks_embeddings[0]))  # should be numpy.ndarray
```

```python
import chromadb
import uuid
import numpy as np

path="./embeddings_db"

client = chromadb.PersistentClient(path=path) 

# --- 3. Get or Create a Collection ---
# A collection is like a table in a database, used to store a specific set of embeddings.
collection = client.get_or_create_collection(name="embedding_collection")

# --- 4. Prepare Data for Insertion ---
# ChromaDB requires a unique ID, the embedding, metadata, and the original document for each entry.
ids = [str(uuid.uuid4()) for _ in final_chunks]
documents = final_chunks
embeddings_to_insert = chunks_embeddings # Convert numpy array to a standard list

# --- 5. Add the Data to the Collection ---
# This command inserts all your data into the database.
# The PersistentClient automatically saves the data to the disk.
collection.add(
    ids=ids,
    embeddings=embeddings_to_insert,

    documents=documents
)

print(f"Successfully added {collection.count()} chunks to the '{collection.name}' collection.")
print(f"The data has been automatically saved to the '{path}' directory.")
```
